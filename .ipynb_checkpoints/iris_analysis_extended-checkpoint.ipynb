# Iris Species — Extended Analysis & Modeling

This notebook:

- Loads the provided dataset
- Performs extended exploratory data analysis and charts
- Trains multiple models with cross-validation
- Performs hyperparameter tuning (GridSearchCV)
- Selects and saves the best model
- Provides **deployment-ready** prediction code (function + Flask example)

**Notes:**
- Uses `matplotlib` for plots (no seaborn).
- Assumes the dataset is at `/mnt/data/Iris_Data.csv`. If your file has a different name/path, edit the path in the cell below.

```python
# Imports & load data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import joblib
```

```python
# Load dataset
data_url = r'C:\Users\hkaluuma\Desktop\Machine_learning\Iris_Data.csv'
df = pd.read_csv(data_url)
df.head()
```

```python
# Basic Exploratory Data Analysis (EDA)
print('Shape:', df.shape)
print('\nColumns:', df.columns.tolist())
print('\nMissing values:\n', df.isna().sum())

display(df.describe())
```

```python
# Histograms for each numeric feature
numeric_cols = ['sepal_length','sepal_width','petal_length','petal_width']

plt.close('all')
for c in numeric_cols:
    plt.figure()
    df[c].hist()
    plt.title(f'Histogram of {c}')
    plt.xlabel(c)
    plt.ylabel('Count')
    plt.show()

# Boxplots
plt.figure()
df[numeric_cols].plot(kind='box', subplots=True, layout=(1,4), figsize=(12,4))
plt.suptitle('Boxplots of numeric features')
plt.show()
```

```python
# Scatter matrix (pairwise relationships)
plt.figure(figsize=(8,8))
scatter_matrix(df[numeric_cols], diagonal='hist', figsize=(10,10))
plt.suptitle('Scatter matrix of features')
plt.show()
```

```python
# Correlation matrix heatmap using imshow
corr = df[numeric_cols].corr()
plt.figure()
plt.imshow(corr, interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=45)
plt.yticks(range(len(numeric_cols)), numeric_cols)
plt.title('Feature correlation matrix')
for i in range(len(numeric_cols)):
    for j in range(len(numeric_cols)):
        plt.text(j, i, f'{corr.iloc[i,j]:.2f}', ha='center', va='center', color='w')
plt.show()
```

```python
# Encode species and split data
le = LabelEncoder()
df['Species_enc'] = le.fit_transform(df['species'])

X = df[numeric_cols].values
y = df['Species_enc'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

print('Training samples:', X_train.shape[0], 'Test samples:', X_test.shape[0])
print('Classes:', le.classes_)
```

```python
# Define pipelines for several models and compare with cross-validation
models = {
    'LogisticRegression': Pipeline([('scaler', StandardScaler()), ('clf', LogisticRegression(max_iter=1000))]),
    'SVC': Pipeline([('scaler', StandardScaler()), ('clf', SVC())]),
    'RandomForest': Pipeline([('scaler', StandardScaler()), ('clf', RandomForestClassifier())]),
    'GradientBoosting': Pipeline([('scaler', StandardScaler()), ('clf', GradientBoostingClassifier())])
}

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

cv_results = {}
for name, pipeline in models.items():
    scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring='accuracy')
    cv_results[name] = scores
    print(f'{name}: mean accuracy = {scores.mean():.4f}, std = {scores.std():.4f}')
```

```python
# Hyperparameter tuning using GridSearchCV for two candidate models
param_grids = {
    'RandomForest': {
        'clf__n_estimators': [50, 100, 200],
        'clf__max_depth': [None, 3, 5, 10],
        'clf__min_samples_split': [2, 5, 10]
    },
    'SVC': {
        'clf__C': [0.1, 1, 10],
        'clf__kernel': ['rbf','linear'],
        'clf__gamma': ['scale','auto']
    }
}

best_models = {}
for name in ['RandomForest','SVC']:
    pipe = Pipeline([('scaler', StandardScaler()), ('clf', models[name].named_steps['clf'])])
    grid = GridSearchCV(pipe, param_grids[name], cv=cv, scoring='accuracy', n_jobs=-1)
    grid.fit(X_train, y_train)
    print('\n', name, 'best params:', grid.best_params_)
    print(name, 'best CV accuracy:', grid.best_score_)
    best_models[name] = grid.best_estimator_
```

```python
# Evaluate tuned models on the hold-out test set
for name, model in best_models.items():
    preds = model.predict(X_test)
    acc = accuracy_score(y_test, preds)
    print('\nModel:', name)
    print('Test accuracy:', acc)
    print(classification_report(y_test, preds, target_names=le.classes_))
    cm = confusion_matrix(y_test, preds)
    print('Confusion matrix:\n', cm)

    # Plot confusion matrix as image
    plt.figure()
    plt.imshow(cm, interpolation='nearest')
    plt.title(f'Confusion matrix - {name}')
    plt.colorbar()
    tick_marks = np.arange(len(le.classes_))
    plt.xticks(tick_marks, le.classes_, rotation=45)
    plt.yticks(tick_marks, le.classes_)
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            plt.text(j, i, cm[i, j], ha='center', va='center', color='w')
    plt.xlabel('Predicted label')
    plt.ylabel('True label')
    plt.show()
```

```python
# Choose the overall best model based on test accuracy and save it
evaluations = {}
for name, model in best_models.items():
    preds = model.predict(X_test)
    evaluations[name] = accuracy_score(y_test, preds)

best_name = max(evaluations, key=evaluations.get)
best_model = best_models[best_name]
print('Best model selected:', best_name, 'with test accuracy =', evaluations[best_name])

# Save the trained best model and label encoder for deployment
joblib.dump({'model': best_model, 'label_encoder': le}, 'C:\\Users\\hkaluuma\\Desktop\\Machine_learning\\best_iris_model.joblib')
print('Saved best model to C:\Users\hkaluuma\Desktop\Machine_learning\best_iris_model.joblib')
```

```python
# Deployment-ready prediction helpers

def predict_species(sample, model_path='/mnt/data/best_iris_model.joblib'):
    """Predict species from a sample dict or list/array of features.

    sample can be:
    - dict: {'SepalLengthCm': ..., 'SepalWidthCm': ..., 'PetalLengthCm': ..., 'PetalWidthCm': ...}
    - list/array: [SepalLengthCm, SepalWidthCm, PetalLengthCm, PetalWidthCm]
    """
    data = joblib.load(model_path)
    model = data['model']
    le = data['label_encoder']

    if isinstance(sample, dict):
        features = [sample[c] for c in ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
    else:
        features = list(sample)
    arr = np.array(features).reshape(1, -1)
    pred_enc = model.predict(arr)[0]
    pred_label = le.inverse_transform([pred_enc])[0]
    return pred_label

# Example usage:
print('Example prediction:')
example = {'SepalLengthCm':5.1,'SepalWidthCm':3.5,'PetalLengthCm':1.4,'PetalWidthCm':0.2}
print('Input:', example)
print('Predicted species:', predict_species(example))

# Simple Flask app skeleton for deployment (save as app.py)
flask_app_code = r"""
from flask import Flask, request, jsonify
import joblib
import numpy as np

app = Flask(__name__)
model_bundle = joblib.load('/mnt/data/best_iris_model.joblib')
model = model_bundle['model']
le = model_bundle['label_encoder']

@app.route('/predict', methods=['POST'])
def predict():
    payload = request.json
    # Expect JSON with either 'features' (list) or individual keys
    if 'features' in payload:
        arr = np.array(payload['features']).reshape(1, -1)
    else:
        arr = np.array([payload[k] for k in ['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]).reshape(1, -1)
    pred = model.predict(arr)[0]
    label = le.inverse_transform([pred])[0]
    return jsonify({'species': label})

if __name__ == '__main__':
    app.run(host='0.0.0.0', port=5000)
"""

with open('/mnt/data/flask_app_example.py','w') as f:
    f.write(flask_app_code)
print('\nWrote Flask example to /mnt/data/flask_app_example.py')
```
